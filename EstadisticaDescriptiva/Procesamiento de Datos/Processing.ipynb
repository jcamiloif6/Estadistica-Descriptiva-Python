```python
import pandas as pd

data = {"nombre" : ["Mariana", "Ana", "Elsa", "Gustavo",
                    "Pedro", "Raul", "Carlos", "Jose", "Luis"],
        "saldo" : [10000, 8000, 9000, 2000, 2100, 
                  12000, 5000, 10000, 200],
        "pais" : ["Argentina", "Bolivia", "Chile", "Colombia", 
                 "Costa Rica", "Ecuador", "Mexico", "Peru", "Peru"]}
data = pd.DataFrame(data)
data["pais"] = data["pais"].astype("category")
data.info()
data
```

## <font color='#FF9933'>What Not To Do</font>

```python
datos_sesgados = data.copy()

swap = {"Argentina" : 1,
       "Bolivia" : 2,
       "Chile" : 3,
       "Colombia" : 4,
       "Costa Rica" : 5,
       "Ecuador" : 6,
       "Mexico" : 7,
       "Peru" : 8}

datos_sesgados["pais"].replace(swap, inplace=True)
datos_sesgados
```

## <font color='#FF9933'>Codificacion de Categorias One-hot</font>

```python
from sklearn.preprocessing import OneHotEncoder

coder = OneHotEncoder()

codif = coder.fit_transform(data[["pais"]])

#print(type(codif))
#print(codif)
#print(codif.toarray())

new_cols = pd.DataFrame(codif.toarray(), columns=coder.categories_)
print(new_cols)

data = pd.concat([data, new_cols], axis="columns")
data
```

```python
data.drop("pais", axis="columns", inplace=True)
data
```
