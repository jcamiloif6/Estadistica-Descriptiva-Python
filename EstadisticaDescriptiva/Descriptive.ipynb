## <font color='#FF9933'>Escalamiento o Normalización de Datos Numéricos</font>

```python
import timeit
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model

X, y = datasets.load_diabetes(return_X_y=True)
raw = X[:, None, 2]
```

### <font color='#5FC0C3'>Escalamiento min-max</font>

```python

max_raw = max(raw)
min_raw = min(raw)
scaled = (2*raw - min_raw - max_raw)/(max_raw - min_raw)

fig, ax = plt.subplots(2,1, sharex=True)

ax[0].hist(raw)
ax[1].hist(scaled)
```

```python
# Modelo de entrenamiento
def train_raw():
    linear_model.LinearRegression().fit(raw,y)

def train_scaled():
    linear_model.LinearRegression().fit(scaled,y)
```

```python
raw_time = timeit.timeit(train_raw, number=100)
scaled_time = timeit.timeit(train_scaled, number=100)

print("train raw: ", raw_time)
print("train scaled: ", scaled_time)
```

### <font color='#5FC0C3'>Escalamiento Z-Score</font>

```python
import numpy as np

mean_raw = np.mean(raw)
stdv_raw = np.std(raw)

scaled_zscore = (raw-mean_raw)/stdv_raw

fig, ax = plt.subplots(2,1, sharex=True)

ax[0].hist(raw)
ax[1].hist(scaled_zscore)
plt.show()
```

```python
# Entrenamiento 
def train_raw():
    linear_model.LinearRegression().fit(raw,y)

def train_scaled():
    linear_model.LinearRegression().fit(scaled_zscore,y)
```

```python
raw_time = timeit.timeit(train_raw, number=100)
scaled_time = timeit.timeit(train_scaled, number=100)

print("train raw: ", raw_time)
print("train scaled: ", scaled_time)
```

# <font color='#FF9933'>Reducción de dimensionalidad con PCA</font>

## <font color='#5FC0C3'>Paso 1: Estandarizar los datos</font>

```python
import pandas as pd
import matplotlib.pyplot as plt

datos = {"ingreso" : [9136.81, 11709.94, 12416.79, 4875.73, 10065.03,
                      11755.33, 7401.78, 10291.38, 12002.42, 11787.76,
                      13347.65, 7676.76, 9808.56, 11556.47, 7875.49,
                      12486.19, 9242.40, 6867.09, 12946.25, 9167.36],
         
         "horas_trabajadas" : [40.38, 41.55, 41.91, 40.18, 41.69,
                               42.32, 41.57, 41.26, 42.01, 41.46,
                               41.71, 41.36, 40.82, 41.22, 41.11,
                               41.07, 41.20, 40.61, 41.52, 41.26]}

datos = pd.DataFrame(datos)


fig = plt.figure(figsize=(7,7))
plt.scatter(datos["ingreso"], datos["horas_trabajadas"],
            marker="8", s=550, color="purple", alpha=0.5)
plt.xlabel("Ingreso", fontsize=20, color="darkblue")
plt.ylabel("Promedio de horas Trabajadas", fontsize=20, color="darkblue")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.set_title("Datos Originales", fontsize=15)
ax1.scatter(datos["ingreso"], datos["horas_trabajadas"],
            marker="8", s=550, color="purple", alpha=0.5)
ax1.set_xlabel("Ingreso", fontsize=15)
ax1.set_ylabel("Promedio de horas trabajadas", fontsize=15)


# Estandariazcion de datos
datos = pd.DataFrame(StandardScaler().fit_transform(datos),
                     columns=["ingreso", "horas_trabajadas"])
ax2.set_title("Datos Estandarizados", fontsize=15)
ax2.scatter(datos["ingreso"], datos["horas_trabajadas"],
            marker="8", s=550, color="purple", alpha=0.5)
ax2.set_xlabel("Ingreso", fontsize=15)
ax2.set_ylabel("Promedio de horas trabajadas", fontsize=15)


plt.show()
```

## <font color='#5FC0C3'>Paso 2: Matriz de Covarianza</font>

```python
print(datos.cov())

plt.figure(figsize=(7,7))
plt.scatter(datos["ingreso"], datos["horas_trabajadas"],
            marker="8", s=550, color="purple", alpha=0.5)
plt.xlabel("Ingreso", fontsize=15)
plt.ylabel("Promedio de horas trabajadas", fontsize=15)

plt.show()
```

## <font color='#5FC0C3'>Paso 3: Eigenvalues & Eigenvectors</font>

```python
from numpy.linalg import eig
valores, vectores = eig(datos.cov())

vector_azul = vectores[:, 0]
vector_rojo = vectores[:, 1]

print(vectores)
print(valores)

print("Vector propio rojo: ", vector_rojo, " valor propio: ", valores[1])
print("Vector propio azul: ", vector_azul, " valor propio: ", valores[0])

plt.figure(figsize=(7,7))
plt.axes().set_aspect("equal")

plt.scatter(datos["ingreso"], datos["horas_trabajadas"],
            marker="8", s=550, color="purple", alpha=0.5)

plt.quiver(0,0,
           vector_azul[0]/abs(vector_azul[0])*valores[0], 
           vector_azul[1]/abs(vector_azul[1])*valores[0],
           color="blue", angles="xy", scale_units="xy", scale=1, width=0.02)

plt.quiver(0,0,
           vector_rojo[0]/abs(vector_rojo[0])*valores[1], 
           vector_rojo[1]/abs(vector_rojo[1])*valores[1],
           color="red", angles="xy", scale_units="xy", scale=1, width=0.02)

plt.xlabel("Ingreso", fontsize=15)
plt.ylabel("Promedio Horas Trabajadas", fontsize=15)


plt.show()
```

## <font color='#5FC0C3'>Paso 4: Proyectar datos</font>

```python
proyectados = pd.DataFrame(datos.values @ vectores.T,
                          columns=["ingreso", "horas_trabajadas"])

plt.figure(figsize=(8,8))
plt.axes().set_aspect("equal")

plt.scatter(proyectados["ingreso"], proyectados["horas_trabajadas"],
            marker="8", s=550, color="purple", alpha=0.5)

plt.scatter(proyectados["ingreso"],[-2]*len(proyectados["horas_trabajadas"]),
            marker="8", s=250, color="red", alpha=0.5)

plt.scatter([-3]*len(proyectados["ingreso"]),proyectados["horas_trabajadas"],
            marker="8", s=250, color="blue", alpha=0.5)

plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA()
datos = pca.fit_transform(datos)

plt.figure(figsize=(9,9))
plt.axes().set_aspect("equal")
plt.scatter(datos[:,0], [2]*datos[:,0].size,
            marker="8", s=250, color="red", alpha=0.5)
plt.scatter(datos[:,1], [1]*datos[:,1].size,
            marker="8", s=250, color="blue", alpha=0.5)

plt.ylim(0.8,2.2)

plt.show()
```
